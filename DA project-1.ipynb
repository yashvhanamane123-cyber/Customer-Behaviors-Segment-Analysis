```python
import pandas as pd
df = pd.read_csv("customer_shopping_behavior_pro1.csv")
df
```

```python
df.columns
```

```python
#creating new column of total_sales by multiplying previous purchases
df['Total_Sales'] = df['Previous Purchases'] * df['Purchase Amount (USD)']
```

```python
df.head(20)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatterplot of Age vs Sales
sns.scatterplot(x='Age', y='Total_Sales', data=df)

# Add title and labels
plt.title('Sales vs Age')
plt.xlabel('Age')
plt.ylabel('Sales')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Total sales per gender
gender_sales = df.groupby('Gender')['Total_Sales'].sum()

# Bar chart
gender_sales.plot(kind='bar', color=['pink', 'salmon'])
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
#snakecasse 
df.columns = df.columns.str.lower() #will convert the string in lowercase
df.columns = df.columns.str.replace(' ', '_')#will convert the spaces into underscore
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
#creating new column age_group 
import pandas as pd
labels = ['Young','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)

#here we have 4 labels that we defined. so that is what we used (pd.qcut)function to divide the column histroy of 
#age into 4 divisions and to assign the labels.
```
